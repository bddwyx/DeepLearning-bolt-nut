```python
#解压数据集
#!cd load/data/data6045/ && unzip -qo lslm.zip && unzip -qo lslm-test.zip

#####预处理
import os
import shutil
import codecs


train_path = 'load/data/data6045/lslm'
with codecs.open('load/data/data6045/train.txt', 'w') as f:
    with codecs.open('load/data/data6045/lslm/train.txt') as r:
        for line in r:
            f.write(train_path + '/' + line)


eval_path = 'load/data/data6045/lslm-test'
with codecs.open('load/data/data6045/eval.txt', 'w') as f:
    with codecs.open('load/data/data6045/lslm-test/eval.txt') as r:
        for line in r:
            f.write(eval_path + '/' + line)

if os.path.exists('load/data/data6045/lslm/label_list.txt'):
    shutil.move('load/data/data6045/lslm/label_list.txt', 'load/data/data6045/label_list.txt')
if os.path.exists('load/data/data6045/lslm/label_list'):
    shutil.move('load/data/data6045/lslm/label_list', 'load/data/data6045/label_list')
    
# label_list.txt 每一行一个类别，类别编号\t类别名字
# train.txt 每一行一个样本，图片路径\t{"bbox":{"value":"bolt","coordinate":[[1014.261,424],[1181.218,707.826]]}\t{...}... 注意，每个字段值之间用\t分割。坐标中有两组坐标，分别表示左上角和右下角
# eval.txt 格式通 train.txt
```
